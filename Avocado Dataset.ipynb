ANALYSIS

```python
#describe the data
av.describe()
```

```python
#checking the column names
av.columns
```

```python
#change date datatype from object to datetime
av['Date'] = pd.to_datetime(av['Date'])
```

```python
av.info()
```

```python
av1 = av.groupby('Year')[['Total Volume']].sum()
av1
```

```python
av1.plot.bar(color = 'purple')
```

```python
av2 = av.groupby('Region')[['Total Volume']].mean().sort_values(by = 'Total Volume', ascending = False)
av2
```

```python
av2.plot.bar(color = 'green', figsize = (9,9))
```

```python
#removing TotalUS and conducting new analysis
av3 = av[av['Region'] != 'TotalUS']
av3.head()
```

```python
av4 = av3.groupby('Region')[['Total Volume']].mean().sort_values(by = 'Total Volume', ascending = False)
av4
```

```python
av4.plot.bar(color = 'orange', figsize = (9,9))
```

```python
#creating a new column for quarter
av['Quarter'] = av['Date'].dt.quarter
av.head()
Quarter = {1:'Jan-Mar', 2:'Apr-Jun', 3:'Jul-Sept', 4:'Oct-Dec'}
av['Quarter'].replace(Quarter, inplace = True)
```

```python
av.head()
```

```python
A = av['Region'].isin(['West', 'California', 'SouthCentral', 'Northeast', 'Southeast', 'GreatLakes'])
B = av['Year'] == 2015

av5 = av[A & B]
av5
```

```python
av6 = av5.groupby(['Quarter'])[['Total Volume']].mean().sort_values(by = 'Total Volume', ascending = False)
av6
```

```python
av6.plot.line()
```

```python
av6.plot.bar(color = 'red')
```

```python
av7 = av5.groupby(['Region'])[['Total Volume']].mean().sort_values(by = 'Total Volume', ascending = False)
av7
```

```python
av7.plot.bar(color = 'black')
```
